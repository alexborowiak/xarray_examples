This notebook is just for the creation of a small amount of sample data to be used. Please make sure in your google drive home page you have a folder called "Colab Notebooks".

```python
import numpy as np
import xarray as xr
```

```python
ds = xr.tutorial.open_dataset( "air_temperature", chunks={"lat": "auto","lon": 25,"time": -1,})
ds
```

```python
# Creating some small subsets to be saved off
ds1 = ds.isel(time=slice(0,20)).fillna(0)
ds2 = ds.isel(time=slice(21, 40)).fillna(0)
ds3 = ds.isel(time=slice(41,60)).fillna(0)
ds4 = ds.isel(time=slice(61, 80)).fillna(0)
```

```python
# Saving the data to the Colab Notebooks folder
ds1.to_netcdf('ds1900.nc')
ds2.to_netcdf('ds1901.nc')
ds3.to_netcdf('ds1902.nc')
ds4.to_netcdf('ds1904.nc')
```
